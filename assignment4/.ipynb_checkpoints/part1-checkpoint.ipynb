```python
import pandas as pd
from pandas import DataFrame
dtype_dict = {'bathrooms':float, 'waterfront':int, 'sqft_above':int, 'sqft_living15':float, 'grade':int, 'yr_renovated':int, 'price':float, 'bedrooms':float, 'zipcode':str, 'long':float, 'sqft_lot15':float, 'sqft_living':float, 'floors':float, 'condition':int, 'lat':float, 'date':str, 'sqft_basement':int, 'yr_built':int, 'id':str, 'sqft_lot':int, 'view':int}

sales = pd.read_csv('kc_house_data.csv', dtype=dtype_dict)
sales = sales.sort(['sqft_living','price'])
```

```python
def polynomial_sframe(feature, degree):
    # assume that degree >= 1
    # initialize the SFrame:
    poly_sframe = DataFrame()
    # and set poly_sframe['power_1'] equal to the passed feature
    poly_sframe['power_1'] = feature
    # first check if degree > 1
    if degree > 1:
        # then loop over the remaining degrees:
        # range usually starts at 0 and stops at the endpoint-1. We want it to start at 2 and stop at degree
        for power in range(2, degree+1): 
            # first we'll give the column a name:
            name = 'power_' + str(power)
            # then assign poly_sframe[name] to the appropriate power of feature
            poly_sframe[name] = map(lambda x: x**power, feature)
    return poly_sframe
```

```python
l2_small_penalty = 1.5e-5
```

```python
from sklearn import linear_model
import numpy as np

poly15_data = polynomial_sframe(sales['sqft_living'], 15) # use equivalent of `polynomial_sframe`
model = linear_model.Ridge(alpha=l2_small_penalty, normalize=True)
model.fit(poly15_data, sales['price'])
```

```python
#Quiz Question: Whatâ€™s the learned value for the coefficient of feature power_1?
```

```python
print "learned value for the coefficient of feature power_1 is " + str(model.coef_[0])
```

```python
set_1 = pd.read_csv('wk3_kc_house_set_1_data.csv', dtype=dtype_dict)
set_2 = pd.read_csv('wk3_kc_house_set_2_data.csv', dtype=dtype_dict)
set_3 = pd.read_csv('wk3_kc_house_set_3_data.csv', dtype=dtype_dict)
set_4 = pd.read_csv('wk3_kc_house_set_4_data.csv', dtype=dtype_dict)
```

```python
l2_small_penalty=1e-9
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def fitAndPlot(data, l2_penalty):
    poly_data = polynomial_sframe(data['sqft_living'], 15)
    features = poly_data.columns
    model = linear_model.Ridge(alpha=l2_penalty, normalize=True)
    model.fit(poly_data, data['price'])
    plt.plot(poly_data['power_1'],data['price'],'.',
        poly_data['power_1'], model.predict(poly_data[features]),'-')
    return model
```

```python
model1 = fitAndPlot(set_1, l2_small_penalty)
model2 = fitAndPlot(set_2, l2_small_penalty)
model3 = fitAndPlot(set_3, l2_small_penalty)
model4 = fitAndPlot(set_4, l2_small_penalty)
```

```python
train_valid_shuffled = pd.read_csv('wk3_kc_house_train_valid_shuffled.csv', dtype=dtype_dict)
test = pd.read_csv('wk3_kc_house_test_data.csv', dtype=dtype_dict)
```

```python
n = len(train_valid_shuffled)
k = 10 # 10-fold cross-validation

for i in xrange(k):
    start = (n*i)/k
    end = (n*(i+1))/k-1
    print i, (start, end)
```

```python
def k_fold_cross_validation(k, l2_penalty, data, output):
    sumErr = 0
    for i in xrange(k):
        start = (n*i)/k
        end = (n*(i+1))/k-1
        train = data[0:start].append(data[end+1:n])
        validate = data[start:end+1]
        train_label = output[0:start].append(output[end+1:n])
        validate_label = output[start:end+1]
        model = linear_model.Ridge(alpha=l2_penalty, normalize=True)
        model = model.fit(train, train_label)
        error = ((model.predict(validate) - validate_label) *(model.predict(validate) - validate_label)).sum()
        sumErr = sumErr + error
    
    return sumErr/k
```

```python
cv_data = polynomial_sframe(train_valid_shuffled['sqft_living'], 15)
```

```python
minError = k_fold_cross_validation(10, 1000, cv_data, train_valid_shuffled['price'])
bestP = 1000
for l2_penalty in np.logspace(4, 9, num=13):
    error = k_fold_cross_validation(10, l2_penalty, cv_data, train_valid_shuffled['price'])
    if (error < minError):
        minError = error
        bestP = l2_penalty
(bestP, minError)
    
```
